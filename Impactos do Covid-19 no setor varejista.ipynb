# Impactos do COVID-19 nos setores do varejo

```python
import pandas as pd
from pandas_datareader import data as wb
import numpy as np
import matplotlib.pyplot as plt
```

# Moda

São constituídas as empresas como o grupo Guarapes, C&A, Marisa, Hering e Renner. Sendo as empresas que mais são afetadas pelo fechado de lojas físicas.

```python
acoes=['GUAR3.SA','CEAB3.SA','AMAR3.SA','HGTX3.SA','LREN3.SA','^BVSP']
```

```python
moda=pd.DataFrame()
for i in acoes:
    moda[i]=wb.DataReader(i, data_source='yahoo', start='2020-2-1') ['Adj Close']
```

```python
moda.head()
```

```python
moda=(moda/moda.iloc[0])*100
```

# Gráfico de comparação

```python
plt.style.use('dark_background')
plt.figure(figsize=(15,8))
plt.plot(moda['GUAR3.SA'], label='Riachuelo', color='white');
plt.plot(moda['CEAB3.SA'], label='C&A', color='blue');
plt.plot(moda['AMAR3.SA'], label='Marisa', color='deeppink');
plt.plot(moda['HGTX3.SA'], label='Hering', color='midnightblue');
plt.plot(moda['LREN3.SA'], label='Renner', color='red');
plt.plot(moda['^BVSP'], label='Ibovespa', linestyle='--', linewidth=3, color='yellow');
plt.title('Varejistas de moda X Covid-19')
plt.xlabel('Data')
plt.ylabel('Performance')
plt.legend()
plt.show()
```

Setor foi bastante afetado, pois mesmo tendo vendas pela plataforma digital, entretanto, há uma grande dependência das lojas físicas.
