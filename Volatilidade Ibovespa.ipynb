Iniciaremos importando as bibliotecas necessárias:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as wb
```

A partir disso iremos encontrar os dados a partir da API do yahoo finance, sendo os fechamentos do Ibovespa desde o dia 18/02. 

```python
ibov=wb.DataReader('^BVSP', data_source='yahoo', start='2020-2-18')['Adj Close']
ibov.tail()
```

Iremos reproduzir o gráfico do Ibovespa do dia 18/02 ao 03/04.

```python
plt.plot(ibov)
plt.xlabel('Data')
plt.ylabel('Pontos')
plt.title('Ibovespa')
plt.xticks(rotation=30)
```

Agora iremos calcular a taxa de retorno para encontrarmos a volatilidade no período analisado.

```python
tx_return=np.log(ibov/ibov.shift(1))
days=len(tx_return[1:])
vol=tx_return[1:].std()*days**0.5
```

Com isso encontramos encontramos o que estávamos procurando:

```python
vol=vol*100
print('Volatilidade: %.2f%%' %vol)
```
